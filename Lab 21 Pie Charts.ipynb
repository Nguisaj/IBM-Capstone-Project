```python
Task 1: Visualizing Data Composition with Pie Charts
1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With
In the survey data, the DatabaseWantToWorkWith column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.
```

```python
# Step 1: Analyze the 'DatabaseWantToWorkWith' column
# We'll first count the occurrences of each database in the column.

# Count the occurrences of each database
db_counts = df['DatabaseWantToWorkWith'].value_counts()

# Get the top 5 most popular databases
top_5_databases = db_counts.head(5)

# Step 2: Create a pie chart
# Visualizing the top 5 databases
plt.figure(figsize=(8, 6))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Top 5 Databases Respondents Want to Work With")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Use the correct column name with a space
ai_tools = df['AIToolCurrently Using'].dropna()

# Split multiple tools in one cell into separate entries
all_tools = ai_tools.str.split(';').explode()

# Count the frequency of each tool and get the top 5
top_tools = all_tools.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_tools, labels=top_tools.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Tools Currently Used for AI Development')
plt.axis('equal')
plt.show()



```

```python
3.3 Pie Chart for Preferred Web Frameworks
The WebframeWantToWorkWith column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Drop missing values in the relevant column
web_frameworks = df['WebframeWantToWorkWith'].dropna()

# Split multiple responses per row
all_frameworks = web_frameworks.str.split(';').explode()

# Count frequency of each framework
top_frameworks = all_frameworks.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_frameworks, labels=top_frameworks.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Preferred Web Frameworks')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
3.4 Pie Chart for Most Desired Embedded Technologies
Using the EmbeddedWantToWorkWith column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Clean and prepare the data
embedded_tech = df['EmbeddedWantToWorkWith'].dropna()
all_embedded = embedded_tech.str.split(';').explode()
top_embedded = all_embedded.value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_embedded, labels=top_embedded.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most Desired Embedded Technologies')
plt.axis('equal')
plt.show()
```
